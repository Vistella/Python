```python
#------------------------------------------
#--- Author: Pradeep Singh
#--- Date: 20th January 2017
#--- Version: 1.0
#--- Python Ver: 2.7
#--- Details At: https://iotbytes.wordpress.com/store-mqtt-data-from-sensors-into-sql-database/
#------------------------------------------

import paho.mqtt.client as mqtt
import struct
#import matplotlib.pyplot as plt

#power = []

# MQTT Settings 
MQTT_Broker = "192.168.0.29"
MQTT_Port = 1883
Keep_Alive_Interval = 45
MQTT_Topic = "#"

#Subscribe to all Sensors at Base Topic
def on_connect(client, userdata, flags, rc):
  mqttc.subscribe(MQTT_Topic, 0)
  print("Connected")

#Save Data into DB Table
def on_message(mosq, obj, msg):
	# This is the Master Call for saving MQTT Data into DB
	# For details of "sensor_Data_Handler" function please refer "sensor_data_to_db.py"
	decodedPayload = msg.payload.decode("utf-8")
	print("MQTT Topic: " + msg.topic + ": " + decodedPayload)
	sensor_Data_Handler(msg.topic, decodedPayload)

def on_subscribe(mosq, obj, mid, granted_qos):
    pass



mqttc = mqtt.Client()

# Assign event callbacks
mqttc.on_message = on_message
mqttc.on_connect = on_connect
mqttc.on_subscribe = on_subscribe

# Connect
mqttc.connect(MQTT_Broker, int(MQTT_Port), int(Keep_Alive_Interval))

# Continue the network loop
mqttc.loop_forever()

```

```python
#------------------------------------------
#--- Author: Pradeep Singh
#--- Date: 20th January 2017
#--- Version: 1.0
#--- Python Ver: 2.7
#--- Details At: https://iotbytes.wordpress.com/store-mqtt-data-from-sensors-into-sql-database/
#------------------------------------------
from datetime import date
import json
#import sqlite3
import psycopg2
from psycopg2 import Error

# SQLite DB Name
#DB_Name =  "Tests"

#===============================================================
# Database Manager Class

class DatabaseManager():
    def __init__(self):       
        # Connect to an existing database
        self.conn = psycopg2.connect(user="ashkanahmadidastgerdi", password="", host="localhost",port="6543",database="ashkanahmadidastgerdi")
        self.cursor = self.conn.cursor()
    def add_del_update_db_record(self, sql_query, args=()):
        # SQL query to create a new table
        #create_table_query = "INSERT INTO production.sensor_reading (id, panel_id, created_at, temperature, humidity) VALUES(default,'1', now(),?,?)", [Temperature, Humidity]
        # Execute a command: this creates a new table
        self.cursor.execute(sql_query,args)
        #cursor.execute(create_table_query)
        self.conn.commit()
        print("Table created successfully in PostgreSQL ")
        return

    def __del__(self):
        self.cursor.close()
        self.conn.close()
        print("PostgreSQL connection is closed")

#===============================================================
# Functions to push Sensor Data into Database

# Function to save Temperature to DB Table
def DHT22_Temp_Data_Handler(jsonData):
	#Parse Data 
	#json_Dict = json.loads(jsonData)
	#SensorID = json_Dict['Sensor_ID']
	#Data_and_Time = json_Dict['Date']
	#Temperature = json_Dict['Temperature']
	#SensorID = "Temp"
	#Data_and_Time = "today"
	Temperature = jsonData
	
	#Push into DB Table
	dbObj = DatabaseManager()
	dbObj.add_del_update_db_record("INSERT INTO production.sensor_reading (created_at, temperature) VALUES(now(),5)", [str(Temperature)])
	#dbObj.add_del_update_db_record("INSERT INTO production.sensor_reading (created_at, temperature) VALUES(now(),0)")
	del dbObj
	print("Inserted Temperature Data into Database.")
	print("")

# Function to save Humidity to DB Table
def DHT22_Humidity_Data_Handler(jsonData):
	#Parse Data 
	#json_Dict = json.loads(jsonData)
	#SensorID = json_Dict['Sensor_ID']
	#Data_and_Time = json_Dict['Date']
	Humidity = jsonData
	
	#Push into DB Table
	dbObj = DatabaseManager()
	dbObj.add_del_update_db_record("INSERT INTO production.sensor_reading (created_at, humidity) VALUES(now(),5)", [str(Humidity)])
	del dbObj
	print("Inserted Humidity Data into Database.")
	print("")


#===============================================================
# Master Function to Select DB Funtion based on MQTT Topic

def sensor_Data_Handler(Topic, jsonData):
	if Topic == "Temperature":
		DHT22_Temp_Data_Handler(jsonData)
		
	elif Topic == "Humidity":
		DHT22_Humidity_Data_Handler(jsonData)	

#===============================================================
```
