```python
pricing_const = {}
pricing_output = {}

with open("pricing_constant.csv", "r") as file:
    content = file.read()
    for lines in content.splitlines():
        pricing_const[lines.split(',')[0]] = lines.split(',')[1]

for key,value in zip(pricing_const.keys(),pricing_const.values()):
    pricing_const[key] = float(value)
#     exec(key + "=float(value)")

#non-row optimised
pricing_const['panel_number'] = pricing_const['farm_area']*pricing_const['panel_density']/pricing_const['panel_area']
pricing_output['inverter_cost'] = pricing_const['inverter_unitcost']*pricing_const['panel_number']*pricing_const['panel_ratedpower']
pricing_output['wire_cost'] = 0
pricing_output['land_cost'] = pricing_const['farm_area']*pricing_const['land_unitcost']
pricing_output['panel_cost'] = pricing_const['panel_number']*pricing_const['panel_unitcost']

pricing_output['total_cost'] = sum(pricing_output.values())

with open("pricing_output.csv","w") as file:
    for key,value in pricing_output.items():
        file.write("{},{}\n" .format(key,value))
        
```
